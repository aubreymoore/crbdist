```python
import pandas as pd
```

```python
df = pd.read_csv('../test.csv', sep='|')
df
```

```python
df.columns
```

```python
# rename columns
df.columns = ['decimalLongitude', 'decimalLatitude', 'Notes', 'Source', 'stateProvince', 'Year','eventDate', 'country']
df
```

```python
# Create new columns
df['id'] = df.index+1
df['islandGroup'] = ''
df['island'] = '' 
df['locality'] = '' 
df['coordinateUncertaintyInMeters'] = 1000
df['georeferenceRemarks'] = ''
df['occurrenceRemarks'] = ''
df
```

```python
df.columns
```

```python
# Reorder columns
df = df.reindex(columns = ['id','eventDate','country','stateProvince','islandGroup','island','locality', 
                      'decimalLatitude','decimalLongitude','coordinateUncertaintyInMeters',
                      'georeferenceRemarks','occurrenceRemarks','Notes','Source'])
df

```

```python
df.to_csv('../DwCA/occurrences_1.csv', sep='|', index=False)
```
